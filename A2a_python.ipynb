LOADING THE DATASET

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("NSSO68.csv")
df.head()
```

SELECTING DEPENDENT & INDEPENDENT VARIABLE TO PERFORM REGRESSION<br><br>
Dependent variable - foodtotal_v<br>
Independent variable - picle_v, sauce_jam_v, etc

```python
# Select dependent and independent variables
Y = df['foodtotal_v']
X = df[['pickle_v', 'sauce_jam_v', 'Othrprocessed_v', 'Beveragestotal_v', 'fruits_df_tt_v', 'fv_tot']]
```

```python
print("Independent variable")
X.head(20)
```

```python
print("Dependent Variable")
Y.head(20)
```

PERFORMING MULTIPLE REGRESSION ANALYSIS

```python
import statsmodels.api as sm

# Add a constant to the model (intercept)
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(Y, X).fit()

# Print the summary of the regression
print(model.summary())
```

REGRESSION DIAGNOSTICS

```python
#RESIDUAL PLOT
import matplotlib.pyplot as plt
import seaborn as sns

# Plot residuals
residuals = model.resid
fitted = model.fittedvalues
```

```python
import scipy.stats as stats

# Q-Q plot
fig = sm.qqplot(residuals, line ='45')
plt.title('Q-Q plot')
plt.show()
```

```python
#Homoscadasticity test
from statsmodels.stats.diagnostic import het_breuschpagan

# Perform the Breusch-Pagan test
bp_test = het_breuschpagan(residuals, X)
labels = ['LM Statistic', 'LM Test p-value', 'F-Statistic', 'F-Test p-value']
print(dict(zip(labels, bp_test)))
```

```python
#Multicollinearity check
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Calculate VIF for each explanatory variable
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns

print(vif)
```

CORRECT THE MODEL IF NECESSARY

```python
# Remove variables with high VIF
X_corrected = X.drop(['Othrprocessed_v'], axis=1)  # Example: Removing one variable

# Fit the corrected model
model_corrected = sm.OLS(Y, X_corrected).fit()

# Print the summary of the corrected regression
print(model_corrected.summary())
```
